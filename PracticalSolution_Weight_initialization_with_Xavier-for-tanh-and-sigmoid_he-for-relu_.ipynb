# Using he normal

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
plt.scatter(df['cgpa'],df['resume_score'],c=df['placed'])
```

```python
X = df.drop('placed',axis=1)
y = df['placed']
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(10,activation='relu',input_dim=2,kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.get_weights()
```

```python
initial_weights = model.get_weights()
```

```python
model.set_weights(initial_weights)
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
X = df[['cgpa','resume_score']].values
y = df['placed'].values
y = y.astype(np.int_)
# Your labels 'y' should also be a NumPy array or a list

# Plot the decision regions
plot_decision_regions(X, y, clf=model, legend=2)

# Add labels and title
# plt.xlabel('Feature 1')
# plt.ylabel('Feature 2')
plt.title('Decision Regions')

# Show the plot
plt.show()
```

# Weight Initialization with Xavier/Glorat

 **Advantages of Xavier Initialization:**

1. Helps prevent vanishing or exploding gradients during training, which can lead to faster convergence.
2. Works well with deep networks, especially those using tanh / sigmoid activations.
3. Provides a good balance between initialization scale and the number of input and output units.

# Using python funtion ---
# xavier_normal_initializer = tf.keras.initializers.GlorotNormal()

```python
df1 = pd.read_csv('placement.csv')
df.head()
```

```python
X = df1.drop('placed',axis=1)
y = df1['placed']
```

```python
xavier_normal_initializer = tensorflow.keras.initializers.GlorotNormal()
```

```python
model = Sequential()

model.add(Dense(10,activation='tanh',input_dim=2))
model.add(Dense(10,activation='tanh'))
model.add(Dense(10,activation='tanh'))
model.add(Dense(10,activation='tanh'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model = Sequential()

model.add(Dense(10,activation='tanh',input_dim=2,kernel_initializer=xavier_normal_initializer))
model.add(Dense(10,activation='tanh',kernel_initializer=xavier_normal_initializer))
model.add(Dense(10,activation='tanh',kernel_initializer=xavier_normal_initializer))
model.add(Dense(10,activation='tanh',kernel_initializer=xavier_normal_initializer))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X,y,epochs=100,validation_split=0.2)
```
