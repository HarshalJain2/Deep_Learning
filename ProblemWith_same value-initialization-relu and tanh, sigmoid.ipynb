# USING RELU AS Activation function in hidden layer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
plt.scatter(df['cgpa'],df['resume_score'],c=df['placed'])
```

```python
X = df.drop('placed',axis=1)
y = df['placed']
```

```python
# hidden layer activation function is relu here
model = Sequential()

model.add(Dense(2,activation='relu',input_dim=2))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
# Set parameters to 0.25
model.get_weights()
```

```python
initial_weights = model.get_weights()
```

```python
initial_weights[0] = np.ones(model.get_weights()[0].shape)*0.25
initial_weights[1] = np.ones(model.get_weights()[1].shape)*0.25
initial_weights[2] = np.ones(model.get_weights()[2].shape)*0.25
initial_weights[3] = np.ones(model.get_weights()[3].shape)*0.25
```

```python
model.set_weights(initial_weights)
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
model.get_weights()
```

NOTE--------here onlu see change in bias b21 but weights are same and zero no training.

```python
from mlxtend.plotting import plot_decision_regions
X = df[['cgpa','resume_score']].values
y = df['placed'].values
y = y.astype(np.int_)
# Your labels 'y' should also be a NumPy array or a list

# Plot the decision regions
plot_decision_regions(X, y, clf=model, legend=2)

# Add labels and title
# plt.xlabel('Feature 1')
# plt.ylabel('Feature 2')
plt.title('Decision Regions')

# Show the plot
plt.show()
```
